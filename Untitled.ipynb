## ***Isolation Forests***- *In-Model Technique*
***For simplicity, from now on, we will use only the transactions dataset.***

```python
# Convert Date to Number
unix_timestamps = []

for date in df_trans['date']:
    datetime_obj = pd.to_datetime(date)
    unix_timestamp = int(datetime_obj.timestamp())
    unix_timestamps.append(unix_timestamp)

# Add the converted dates as a new column
df_trans['unix_timestamp'] = unix_timestamps
df_trans.drop(['date'], axis=1, inplace=True)


# Standardization
df_trans_columns = df_trans.columns
std = StandardScaler()
df_trans = std.fit_transform(df_trans)
df_trans = pd.DataFrame(df_trans, columns = df_trans_columns)
```

```python
model = IsolationForest(n_estimators=100, max_samples=0.5,
                        contamination='auto', max_features=1.0, bootstrap=False, n_jobs=None,
                        verbose=1, random_state=2020)
model.fit(df_trans)
```

```python
Y = model.predict(df_trans)
```

```python
Y_series = pd.Series(Y)
print(Y_series.value_counts())
```

***1: Normal data points.      -1: Anomalies.***

## Explanation by feature importance

```python
X_train_scaled_df = pd.DataFrame(df_trans, columns=df_trans.columns)

shap_sample = X_train_scaled_df.sample(500)
explainer = shap.Explainer(model.predict, shap_sample)
shap_values = explainer(shap_sample)
shap_values.feature_names = X_train_scaled_df.columns.tolist()
shap.plots.beeswarm(shap_values)
```

```python
anomaly_indices = Y_series[Y_series == -1].index
anomaly_example_idx = anomaly_indices[0]
```

```python
shap.plots.bar(explainer(X_train_scaled_df)[anomaly_example_idx])
```
