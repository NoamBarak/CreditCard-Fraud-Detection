## ***Isolation Forests***- *In-Model Technique*
***For simplicity, from now on, we will use only the transactions dataset.***

```python
# Convert Date to Number
unix_timestamps = []

for date in df_trans['date']:
    datetime_obj = pd.to_datetime(date)
    unix_timestamp = int(datetime_obj.timestamp())
    unix_timestamps.append(unix_timestamp)


# Add the converted dates as a new column
df_trans['unix_timestamp'] = unix_timestamps
timestamp_date_map = dict(zip(df_trans['unix_timestamp'], df_trans['date']))
df_trans.drop(['date'], axis=1, inplace=True)



# # Standardization
# df_trans_columns = df_trans.columns
# std = StandardScaler()
# df_trans = std.fit_transform(df_trans)
# df_trans = pd.DataFrame(df_trans, columns = df_trans_columns)
```

## Counter Examples

```python
def display_anomaly_and_counterexamples(selected_anomaly_index):
    # Find Similar Counterexamples
    selected_anomaly = df_trans.iloc[selected_anomaly_index]
    similarities = np.linalg.norm(df_trans - selected_anomaly, axis=1)
    sorted_indices = np.argsort(similarities)
    
    num_counterexamples = 3
    counterexample_indices = []
    
    for idx in sorted_indices:
        if len(counterexample_indices) >= num_counterexamples:
            break
        if predictions[idx] == 1 and idx != selected_anomaly_index:  
            counterexample_indices.append(idx)
    
    from tabulate import tabulate
    
    # Display the anomaly and its counterexamples
    print("Anomaly:")
    anomaly_timestamp = df_trans.iloc[selected_anomaly_index]['unix_timestamp']
    anomaly_date = timestamp_date_map[anomaly_timestamp]
    anomaly_data = df_trans.iloc[selected_anomaly_index].to_frame().transpose()
    anomaly_data.drop('unix_timestamp', axis=1, inplace=True)
    anomaly_data['date'] = anomaly_date
    print(tabulate(anomaly_data, headers='keys', tablefmt='grid', floatfmt=".8f"))
    print(f"\n\n")
    
    print("Counterexamples (Similar to Anomaly but Not Anomalies):")
    for idx in counterexample_indices:
        counterexample_timestamp = df_trans.iloc[idx]['unix_timestamp']
        counterexample_date = timestamp_date_map[counterexample_timestamp]
        counterexample_data = df_trans.iloc[idx].to_frame().transpose()
        counterexample_data.drop('unix_timestamp', axis=1, inplace=True)
        counterexample_data['date'] = counterexample_date
        print(tabulate(counterexample_data, headers='keys', tablefmt='grid', floatfmt=".8f"))
        print(f"\n")
```

```python
anomaly_indices = np.where(predictions == -1)[0]

display_anomaly_and_counterexamples(anomaly_indices[2])
```

*As we can see, **most of the values are the same**: the location of the transaction is pretty consistent, and all the transactions were made on the same day. **The only significant difference is the transaction dollar amount**, and this is probably the reason why this transaction was tagged as an anomaly.*

```python
display_anomaly_and_counterexamples(anomaly_indices[120])
```

*In this example, this transaction was likely tagged as an anomaly due to a **combination of two aspects: the transaction dollar amount and location**. The difference in dollar amount between the anomaly and counterexamples is less significant than in the previous example. However, in this case, there is also a significant difference in location, even though all transactions were made on the same day. This combination of differences likely caused this transaction to be tagged as an anomaly.*
